```python
import pandas as pd 
import seaborn as sns
from sqlalchemy import engine
from pandasql import sqldf
# funcion lambda que hace escribir funciones en SQL mucho mas facil
pysqldf = lambda q:sqldf(q, globals())
```

```python
sns.barplot(data=df, x='Competition', y='Avg. monthly searches', errorbar=None,estimator=sum)
```

```python
df= df.rename(columns={"Avg. monthly searches":"avg_searches","Competition (indexed value)":"competition_index"})
# con este codigo cambiamos los nombres de unas columnas de arriba del cuadro 
```

```python
q = """
select
Keyword,
avg_searches,
competition,
competition_index
from df
where avg_searches >= 500
"""
df_export = pysqldf(q)

# con esto podemo hacer el codigo de SQL pero lo mas importante es la formula final para poder correr como SQL df_export = pysqldf(q)
```

```python
df_export.to_csv("datos_google.csv")
# con esto volvemos a exportar los datos del archivo llamado datos_google 
```
